```python
import os
from itertools import combinations

import numpy as np
import pandas as pd

from ensembler import Ensembler
```

# Load Files

```python
y_val = pd.read_pickle('y_val.pkl')

n_val = len(y_val)

inception_val = pd.read_csv('inception_valid_output.csv', header=None).values[:, 0].astype(np.int)
inception_test = pd.read_csv('inception_output.csv', header=None).values[:, 0].astype(np.int)

lenet_val = pd.read_csv('lenet_valid_output.csv', header=None).values[:, 0].astype(np.int)
lenet_test = pd.read_csv('lenet_output.csv', header=None).values[:, 0].astype(np.int)

vgg16_val = pd.read_csv('vgg16_valid_output.csv', header=None).values[:, 0].astype(np.int)
vgg16_test = pd.read_csv('vgg16_output.csv', header=None).values[:, 0].astype(np.int)

vgg5_val = pd.read_csv('vgg5_valid_output.csv', header=None).values[:, 0].astype(np.int)
vgg5_test = pd.read_csv('vgg5_output.csv', header=None).values[:, 0].astype(np.int)

xception_val = pd.read_csv('xception_valid_output.csv', header=None).values[:, 0].astype(np.int)
xception_test = pd.read_csv('xception_output.csv', header=None).values[:, 0].astype(np.int)

resnet_val = pd.read_csv('resnet_valid_output.csv', header=None).values[:, 0].astype(np.int)
resnet_test = pd.read_csv('resnet_output.csv', header=None).values[:, 0].astype(np.int)

model_names = ['inception', 'lenet', 'vgg16', 'vgg5', 'xception', 'resnet']
val_predictions = np.array([inception_val, lenet_val, vgg16_val, vgg5_val, xception_val, resnet_val])
test_predictions = np.array([inception_test, lenet_test, vgg16_test, vgg5_test, xception_test, resnet_test])
```

# Do Ensembling!

## Instantiate an Ensembler Object

```python
ensembler = Ensembler(y_val, val_predictions, test_predictions, model_names)
```

## Simple Majority Vote

```python
majority_vote = ensembler.make_majority_vote()
```

### Make Submission File

```python
majority_vote['Prediction'].to_csv('majority_vote.csv', index=False)
```

## Get f-scores

```python
ensembler.get_class_wise_fscores()
```

## Do Weighted Ensembling

```python
ensembler.weighted_ensembling(drop_correlated=False)
```
